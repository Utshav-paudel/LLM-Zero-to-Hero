# Creating datasets

```python
from datasets import load_dataset
dataset = load_dataset("yelp_review_full")                            # loading datasets
```

```python
dataset                                                             # checking features of datasets got 65k training sentence, 50k test
```

```python
from transformers import AutoTokenizer
checkpoint  = "bert-base-cased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)                           # using tokenizer for pretrained bert model
def tokenize_function(examples):                                                # creating tokenize function to map
    return tokenizer(examples["text"], padding="max_length", truncation=True)

tokenized_dataset = dataset.map(tokenize_function,batched=True)                 # creating tokenization of datasets using batch to reduce computational complexity
# tokenized_dataset = dataset.map(lambda x :  tokenizer(x["text"], padding = "max_length", truncation=True), batched=True)     , simple using lambda for function : üòÇ
```

```python
# Using collator for dynamic padding of tokenized input, so that all become of same length
from transformers import DataCollatorWithPadding
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```
